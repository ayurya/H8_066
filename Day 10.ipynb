# Uniform Distribution

```python
from scipy.stats import uniform
```

```python
# random numbers from uniform distribution
n=10000
start=10
width=20
data_uniform = uniform.rvs(
    size=n,
    loc=start,
    scale=width
)

ax=sns.distplot(
    data_uniform,
    bins=100,
    kde=True,
    color='skyblue',
    hist_kws={'linewidth':15,'alpha':1})
ax.set(xlabel='Uniform Distribution', ylabel='Frequency')
```

# Confidence Intervals (CI)

Rentang antara dua nilai dimana nilai suatu sample mean tepat berada di tengah-tengahnya. 

```python
import pandas as pd
import numpy as np
df=pd.read_csv('Heart.csv')
df
```

```python
# Construct CI for female population proportion that has heart disease
df['Sex1']=df.sex.replace({1:"Male",0:"Female"}) #buat kolom sex1, berisi male dan female yang tadinya berupa boolean 1 dan 0
dx=df[['target','Sex1']].dropna()
```

```python
dh=pd.crosstab(dx.target,dx.Sex1)
dh
```

```python
#peluang female menderita heart disease
p_fm=dh.loc[1,'Female']/(dh.loc[1,'Female']+dh.loc[0,'Female'])
p_fm
```

```python
#jumlah female
n_fm=dh['Female'].sum()
n_fm
```

```python
#standard error untuk female
se_fm=np.sqrt(p_fm*(1-p_fm)/n_fm)
se_fm
```

```python
#Buat estimasi peluang dengan CI 95% --> z score=1,96
z=1.96
lcb=p_fm-z*se_fm #lower limit  CI
ucb=p_fm+z*se_fm #upper limit  CI
lcb,ucb
```

```python
#atau hitung menggunakan library 'statsmodels'
import statsmodels.api as sm
sm.stats.proportion_confint(n_fm*p_fm,n_fm)
```
