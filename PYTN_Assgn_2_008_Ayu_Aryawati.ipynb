* This dataset uses the financial definition of a building/building unit, for tax purposes. In case a single entity owns the building in question, a sale covers the value of the entire building. In case a building is owned piecemeal by its residents (a condominium), a sale refers to a single apartment (or group of apartments) owned by some individual.

## Hypothesis Testing

Make your Hypothesis Testing below

Function below are graded function. (1 Points)

1. Apakah data di tiap borough memiliki distribusi normal/Gaussian?
2. Apakah terdapat perbedaan harga berdasarkan building category?


```python
# Graded-Funtion Begin

def NormTest(bor):
    from scipy.stats import shapiro
    data=pd.DataFrame(df_clean[(df_clean['BOROUGH']==bor)]['SALE PRICE'])
    
    stat,p=shapiro(data)
    print('stat=%.3f,p=%.3f' % (stat,p))
    if p>0.05:
        print('Probably Gaussian')
    else:
        print('Probably not Gaussian')
# Graded-Funtion End

# Call the function
NormTest(1)
```

### Conclusion:
Apakah data di tiap borough memiliki distribusi normal/Gaussian? Based on Shapiro test, all transaction in each borough is not normally distributed.

```python
# Graded-Funtion Begin

def TTest(bor1,bor2,sampleSize):
    data1=pd.DataFrame(df_clean[(df_clean['BOROUGH']==bor1)]['SALE PRICE']).sample(sampleSize)
    data2=pd.DataFrame(df_clean[(df_clean['BOROUGH']==bor2)]['SALE PRICE']).sample(sampleSize)
    
    from scipy.stats import ttest_rel
    stat,p=ttest_rel(data1,data2)
    print('Borough %r and Borough %r' % (bor1,bor2))
    print('stat=%.3f,p=%.3f' % (stat,p))
    if p>0.05:
        print('Probably the same distribution')
    else:
        print('Probably different distribution')
# Graded-Funtion End

# Call the function
TTest(4,5,2000)
```

### Conclusion
Apakah terdapat perbedaan harga berdasarkan borough? Based on Paired Student's t-Test, harga properti di tiap borough berbeda.
